```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
# Functional Method
plt.plot(x, y)
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('matplotlib graph')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)
```

```python
# Object-oriendted method
fig = plt.figure()
```

```python
# canvas 에서 축 지정
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
axes1.plot(x, y)
axes2.plot(x, y)

#axes.set_xlabel('X')
#axes.set_ylabel('Y')
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')

fig
```

```python
#  사이즈 지정, 해상도 지정
fig = plt.figure(figsize=(8,2), dpi = 60 )
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
fig
```

```python
fig.savefig('my_pic.png', dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x ** 2, label = 'X squared')
ax.plot(x, x ** 3, label = 'X cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

ax.plot( x, y, color = '#FF8C00', linewidth = 0.5, alpha = 0.5,
       linestyle = '--', marker = 'o', markersize = 20,
       markerfacecolor = 'yellow', markeredgewidth = 3, 
        markeredgecolor = 'green') 
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

ax.plot( x, y, color = 'purple', lw = 2, ls = '--' )

ax.set_xlim( [0, 1] )
ax.set_ylim( [0, 2] )
```
